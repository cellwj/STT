```python
import unitvelo as utv
```

```python
velo = utv.config.Configuration()
velo.IROOT = None
velo.FIT_OPTION = '1'
velo.GPU = -1
velo.BASIS = 'xy_loc'
```

```python
import scanpy as sc
adata_mb = sc.read('../data/mouse_brain.h5ad')
adata_mb.var['highly_variable']=True
adata_mb.write('../data/mouse_brain.h5ad')
```

```python
path_to_adata = '../data/mouse_brain.h5ad'
label = 'Region'
adata = utv.run_model(path_to_adata, label, config_file=velo)
```

```python
import scvelo as scv
scv.pl.velocity_embedding_stream(adata, color='Region', basis = 'xy_loc', dpi=100, title='')
```

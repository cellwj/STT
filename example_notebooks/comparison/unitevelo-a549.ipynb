```python
import unitvelo as utv
```

```python
velo = utv.config.Configuration()
velo.IROOT = None
velo.FIT_OPTION = '1'
velo.GPU = -1
velo.BASIS = 'umap'
```

```python
import scanpy as sc
adata = sc.read('../data/A549_TGFB1.h5ad')
sc.pp.neighbors(adata, n_neighbors = 20)
sc.tl.umap(adata)
adata.write('../data/A549_TGFB1.h5ad')
```

```python
path_to_adata = '../data/A549_TGFB1.h5ad'
label = 'time'
adata = utv.run_model(path_to_adata, label, config_file=velo)
```

```python
import scvelo as scv
scv.pl.velocity_embedding_stream(adata, color='time', basis = 'umap', dpi=100, title='')
```
